```python
from sklearn.ensemble import RandomForestClassifier
from model.analyze import analyze_variables
from model.train import (
    cross_validation,
    grid_report,
    load_data,
    random_seed,
)
from model.model import save_model
import random

random.seed(random_seed)
```

```python
train_dataset, test_dataset = load_data("../data/raw_data.csv")
experiment = "randomforest"
case_folder = f"./reports/{experiment}"


def model_path(output_folder: str):
    return f"{output_folder}/model"
```

```python
output_folder = f"{case_folder}/biased"
grid, summary = cross_validation(
    [("model", RandomForestClassifier())],
    train_dataset,
    {
        "model__n_estimators": [10, 25, 50],
        "model__max_features": ["sqrt", "log2"],
        "model__max_depth": [3, 5],
        "model__min_samples_split": [2, 3],
        "model__min_samples_leaf": [1, 2, 3],
    },
    include_all=True,
    include_bias=True,
)
summary
```

```python
model = grid_report(grid, test_dataset, experiment)
save_model(model, model_path(output_folder))
analyze_variables(model, test_dataset, output_folder)
```

```python
output_folder = f"{case_folder}/complete"
grid, summary = cross_validation(
    [("model", RandomForestClassifier())],
    train_dataset,
    {
        "model__n_estimators": [10, 25, 50],
        "model__max_features": ["sqrt", "log2"],
        "model__max_depth": [3, 5, 7],
        "model__min_samples_split": [2, 3, 4],
        "model__min_samples_leaf": [1, 2, 3],
    },
    include_all=True,
)
summary
```

```python
model = grid_report(grid, test_dataset, experiment)
save_model(model, model_path(output_folder))
analyze_variables(model, test_dataset, output_folder)
```

```python
model = grid_report(grid, test_dataset, experiment)
save_model(model, model_path(output_folder))
analyze_variables(model, test_dataset, output_folder)
```
