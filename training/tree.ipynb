```python
from sklearn.tree import DecisionTreeClassifier
from model.train import (
    cross_validation,
    grid_report,
    load_data,
    random_seed,
)
from model.model import save_model
import random

random.seed(random_seed)
```

```python
train_dataset, test_dataset = load_data("../data/raw_data.csv")
experiment = "tree"
case_folder = f"./reports/{experiment}"


def model_path(output_folder: str):
    return f"{output_folder}/model"
```

```python
output_folder = f"{case_folder}/biased"
grid, summary = cross_validation(
    [("model", DecisionTreeClassifier())],
    train_dataset,
    {
        "model__splitter": ["random", "best"],
        "model__max_depth": [3, 5, 7],
        "model__min_samples_split": [2, 3, 4],
        "model__min_samples_leaf": [1, 2, 3],
    },
    include_all=True,
    include_bias=True,
)
summary
```

```python
model = grid_report(grid, test_dataset, experiment)
```

```python
output_folder = f"{case_folder}/complete"
grid, summary = cross_validation(
    [("model", DecisionTreeClassifier())],
    train_dataset,
    {
        "model__splitter": ["random", "best"],
        "model__max_depth": [30, 35, 40],
        "model__min_samples_split": [2, 3, 4],
        "model__min_samples_leaf": [1, 2, 3],
    },
    include_all=True,
)
summary
```

```python
model = grid_report(grid, test_dataset, experiment)
save_model(model, model_path(output_folder))
```

```python
output_folder = f"{case_folder}/simple"
grid, summary = cross_validation(
    [("model", DecisionTreeClassifier())],
    train_dataset,  
    {
        "model__splitter": ["random", "best"],
        "model__max_depth": [5, 8, 11],
        "model__min_samples_split": [2, 3, 4],
        "model__min_samples_leaf": [4, 5, 6],
    },
)
summary
```

```python
model = grid_report(grid, test_dataset, experiment)
save_model(model, model_path(output_folder))
```
