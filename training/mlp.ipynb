```python
from sklearn.neural_network import MLPClassifier
from model.analyze import analyze_variables, print_tree
from model.train import (
    cross_validation,
    grid_report,
    load_data,
    random_seed,
)
from model.model import save_model
import random

random.seed(random_seed)
```

```python
train_dataset, test_dataset = load_data("../data/raw_data.csv")
experiment = "mlp"
case_folder = f"./reports/{experiment}"


def model_path(output_folder: str):
    return f"{output_folder}/model"
```

```python
output_folder = f"{case_folder}/biased"
grid, summary = cross_validation(
    [("model", MLPClassifier())],
    train_dataset,
    {
        "model__hidden_layer_sizes": [(1,), (2,), (3,), (4,), (5,)],
        "model__activation": ['identity'],
        "model__alpha": [1e-4, 1e-3],
        "model__learning_rate_init": [1e-3, 1e-2]
    },
    include_all=True,
    include_bias=True,
)
summary
```

```python
model = grid_report(grid, test_dataset, experiment)
save_model(model, model_path(output_folder))
analyze_variables(model, test_dataset, output_folder)
```
